## read contry & flag data

```python
import pandas as pd 
import numpy as np

df = pd.read_excel('./contry_and_flag.xlsx')
countries = df[df.columns[0]].values
regions = df[df.columns[1]].values
images = df[df.columns[2]].values
columns = df.columns

country_image = dict(zip(countries,images))
```

## read time series of COVID-19

```python
df = pd.read_csv('./csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
columns = df.columns.values

countries1 = df[df.columns[1]].values
dates = df[df.columns[4:]].values
```

```python
countries_set = set(countries1)
dates_TS = np.zeros((len(countries_set),dates.shape[1]),dtype=int)
countries_TS = len(countries_set) * [0]

countries_list = list(countries_set)
for i in range(len(countries_set)):
    item = countries_list[i]
    idx = countries1 == item
    countries_TS[i] = item
    dates_TS[i,:] = np.sum(dates[idx,:],axis=0)
```
